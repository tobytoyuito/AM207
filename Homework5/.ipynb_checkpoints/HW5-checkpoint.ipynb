### Question 1: Warming up with a multivariate normal distribution

Your astronomer friend working on star clusters has come up with a posterior for the probabiliy of finding a star within a cluster in position $(x,y,z)$. This posterior distribution happens to be a 3 dimensional multivariate normal with mean vector (1, 2, 3), variances each equal to 4, and an intraclass correlation
coefficient of 1.2; that is,

$$ 
p(x,y,z) \sim  N \left(  
\left[
\begin{array}{c}
    1  \\ 
    2  \\ 
    3  \\ 
  \end{array} 
\right],
\left[
  \begin{array}{ccc}
    4 & 1.2 & 1.2 \\ 
    1.2 & 4 & 1.2 \\ 
    1.2 & 1.2 & 4 \\  
  \end{array} 
  \right] \right)
$$ 

He comes to you because he wants to sample from this distribution, but he suspects that there is a bug in the multivariate normal function from numpy. You quickly realize that you can help him by using a Gibbs sampler. 

In the following we write $X = [x,y,z]$ and $X_k$ as the vector containing $k$ compontens of $X$, and $X_{-k}$ as the vector containing all the components except for $k$. 

In order to find any conditional distribution $p(X_k|X_{-k})$ of a multivariate normal dsitribution with dimension k, you can partition the covariance matrix as follows:

$$ \Sigma = \left[ \begin{array}{cc} \Sigma_{11} & \Sigma_{12} \\ \Sigma_{21} & \Sigma_{22}  \end{array} \right] $$

where $\Sigma_{11}$ is the submatrix of the covariance matrix obtained by deleting the rows and columnns where the variances of the conditioning variables are located, $\Sigma_{22}$ is the submatrix obtained by deleting the rows and columnns where the variances of the conditioned variables are located, and $\Sigma_{12} = \Sigma_{21}^T$ are the remaining matrices.

For our 3 dimensional problem we get:

$$
\Sigma = 
\left[
  \begin{array}{ccc}
    \sigma_x^2 & \sigma_x \sigma_y & \sigma_x \sigma_z \\ 
    \sigma_y \sigma_x & \sigma_y^2 & \sigma_y \sigma_z \\ 
    \sigma_z \sigma_x & \sigma_z \sigma_y & \sigma_z^2 \\  
  \end{array} 
  \right]
$$

so if we are interested in $p(x|y,z)$ we have $\Sigma_{11} = \left[\sigma_x^2\right], \Sigma_{12} = \Sigma_{21}^T = \left[ \sigma_x \sigma_y \quad \sigma_x \sigma_z \right]$, and $\Sigma_{22} = \left[   \begin{array}{cc}
    \sigma_y^2 & \sigma_y \sigma_z \\ 
    \sigma_z \sigma_y & \sigma_z^2 \\  
  \end{array} \right]$ 

We can do this for all three conditional distributions and because our example is so nicely symmetric we get the same sub-matrices for each of the conditionals:

$$ \Sigma_{11} = \left[ 4 \right] $$

$$ \Sigma_{12} = \Sigma_{21}^T = \left[ 1.2 \quad 1.2 \right] $$

$$ \Sigma_{22}= \left[ \begin{array}{cc} 4 & 1.2 \\ 1.2 & 4 \end{array} \right] $$

Now we can use the standard decomposition of the multivariate normal, given below. The conditional distributions are univariate Normals with mean values given by:

$$ \hat{x} = \mu_x + \Sigma_{12} \Sigma^{-1}_{22}  \left[   \begin{array}{cc}  y-\mu_y \\ z-\mu_z \end{array} \right] $$
$$ \hat{y} = \mu_y + \Sigma_{12} \Sigma^{-1}_{22}  \left[   \begin{array}{cc}  x-\mu_x \\ z-\mu_z \end{array} \right] $$
$$ \hat{z} = \mu_z + \Sigma_{12} \Sigma^{-1}_{22}  \left[   \begin{array}{cc}  x-\mu_x \\ y-\mu_y \end{array} \right] $$

where $\mu_x, \mu_y, \mu_z$ are $1,2,3$ and $\Sigma_{ij}$ represents the elements of the covariance matrix above.
The similarity of the expressions comes from the fact that the matrix is very symmetric. The covariance matrix
of the conditional distributions is given by:

$$ \hat{\Sigma} = \Sigma_{11}-\Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21} $$

(a) Write the Gibbs sampler for this distribution by sampling sequentially from the three conditional distributions $p(x|y,z), p(y|x,z), p(z|x,y)$. Choose a thinning parameter, burnin factor and total number of iterations that allow you to take 10000 non-autocorrelated draws. (_Hint_: Don't just give us the numbers you use, please explain how you chose them)

(b) In contrast to your friend you actually do trust the multivariate normal function from numpy. As you have your Gibbs sampler now already in place you decide to give blockwise sampling a try. The idea is to sample sequentially from the conditional distributions $p(x,y|z)$ and $p(z|x,y)$. Verify that both the componentewise and the blockwise updates are leading to samples from the same distribution.

##Solution##
**(a)**

The three conditional distributions are:
$$p(x|y,z) \sim Normal(\mu_x + \Sigma_{12} \Sigma^{-1}_{22}  \left[   \begin{array}{cc}  y-\mu_y \\ z-\mu_z \end{array} \right], \Sigma_{11}-\Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21})$$
$$p(y|x,z) \sim Normal(\mu_y + \Sigma_{12} \Sigma^{-1}_{22}  \left[   \begin{array}{cc}  x-\mu_x \\ z-\mu_z \end{array} \right], \Sigma_{11}-\Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21})$$
$$p(z|x,y) \sim Normal(\mu_z + \Sigma_{12} \Sigma^{-1}_{22}  \left[   \begin{array}{cc}  x-\mu_x \\ y-\mu_y \end{array} \right], \Sigma_{11}-\Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21})$$

Therefore, I can use univariate normal distribution in numpy to draw $x$, $y$, $z$

```python
# define mu and sigma
mu_x, mu_y, mu_z = 1.0,2.0,3.0
cov_22 = [[4.0, 1.2],[1.2, 4.0]]
cov_12 = [1.2, 1.2]
cov_11 = [4]

sig = (cov_11 - np.dot(cov_12, np.linalg.inv(cov_22)).dot(cov_12))**.5

# Gibbs Sampling
def Gibbs(N, burnin=0):
    x=np.zeros(N+burnin+1)
    y=np.zeros(N+burnin+1)
    z=np.zeros(N+burnin+1)

    #Initialize x, y and z.
    x[0]=np.random.uniform(0, 10)
    y[0]=np.random.uniform(0, 10)
    z[0]=np.random.uniform(0, 10)

    mu = lambda m0, m1, m2, x_1, x_2: m0 + np.dot(cov_12, np.linalg.inv(cov_22)).dot([x_1-m1, x_2-m2])

    for i in xrange(N+burnin):
        mux = mu(mu_x, mu_y, mu_z, y[i], z[i])
        x[i+1] = np.random.normal(mux, sig)

        muy = mu(mu_y, mu_x, mu_z, x[i+1], z[i])
        y[i+1] = np.random.normal(muy, sig)

        muz = mu(mu_z, mu_x, mu_y, x[i+1], y[i+1])
        z[i+1] = np.random.normal(muz, sig)

    return x[1+burnin:], y[1+burnin:], z[1+burnin:]
```

```python
x,y,z=Gibbs(10000)
```

```python
# Geweke tests
def Geweke(trace, intervals, length):
    # take two parts of the chain. 
    # subsample lenght 
    nsl=length
    z =np.empty(intervals)
    for k in np.arange(0, intervals):
        # beg of each sub samples
        bega=k*length
        begb = len(trace)/2 + k*length
        
        sub_trace_a = trace[bega:bega+nsl]
        sub_trace_b = trace[begb:begb+nsl]
        
        theta_a = np.mean(sub_trace_a)
        theta_b = np.mean(sub_trace_b)
        var_a  = np.var(sub_trace_a)
        var_b  = np.var(sub_trace_b)
        
        z[k] = (theta_a-theta_b)/np.sqrt(var_a + var_b)
    
    return z
```

```python
x_geweke = Geweke(x_thin, 50, 100)
y_geweke = Geweke(y_thin, 50, 100)
z_geweke = Geweke(z_thin, 50, 100)

plt.figure(figsize=[12,8])
plt.plot(x_geweke, color='b', label = '$x$')
plt.plot(y_geweke, color='orange', label = '$y$')
plt.plot(z_geweke, color='y', label = '$z$')
plt.hlines([-2, 2], 0,50, color='r')
plt.title('Geweke Test Plot with 100 Segments', fontsize=16)
plt.ylabel('Test Result')
plt.ylim([-2.5,2.5])
plt.xlim([0,50])
plt.xlabel('Segment')
plt.legend(loc = 'best')
plt.show()
```

**(b)**

In order to derive conditional distribution $p(x,y|z)$, I need to calculate $\Sigma_{11}$, $\Sigma_{12}$, and $\Sigma_{22}$

$$ \Sigma_{11} = \left[ \begin{array}{cc} 4 & 1.2 \\ 1.2 & 4 \end{array} \right]$$

$$ \Sigma_{12} = \Sigma_{21}^T = \left[ 1.2 \quad 1.2 \right] $$

$$ \Sigma_{22}=  \left[ 4 \right] $$

$$p(x,y|z) \sim Normal(\left[   \begin{array}{cc}  \mu_x \\ \mu_y \end{array} \right] + \Sigma_{12} \Sigma^{-1}_{22}  [z-\mu_z], \Sigma_{11}-\Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21})$$

As for $p(z|x,y)$, 

$$ \Sigma_{11}^{'} = \left[ 4 \right] $$

$$ \Sigma_{12}^{'} = \Sigma_{21}^{'T} = \left[ 1.2 \quad 1.2 \right] $$

$$ \Sigma_{22}^{'}= \left[ \begin{array}{cc} 4 & 1.2 \\ 1.2 & 4 \end{array} \right] $$

$$p(z|x,y) \sim Normal(\mu_z + \Sigma_{12}^{'} \Sigma^{'-1}_{22}  \left[   \begin{array}{cc}  x-\mu_x \\ y-\mu_y \end{array} \right], \Sigma_{11}^{'}-\Sigma_{12}^{'} \Sigma_{22}^{'-1} \Sigma_{21}^{'})$$

So I can use Gibbs sampling to draw $x,y$ and $z$ sequentially from distributions above.

```python
# define mu and sigma
mu_x, mu_y, mu_z = 1.0,2.0,3.0
cov_22 = [[4.0, 1.2],[1.2, 4.0]]
cov_12 = [1.2, 1.2]
cov_11 = [4.0]

sig_z = (cov_11 - np.dot(cov_12, np.linalg.inv(cov_22)).dot(cov_12))**.5
cov_xy = cov_22 - np.dot(cov_12, cov_12)/4.0

# Gibbs Sampling
def Gibbs(N, burnin=0):
    x=np.zeros(N+burnin+1)
    y=np.zeros(N+burnin+1)
    z=np.zeros(N+burnin+1)

    #Initialize x, y and z.
    x[0]=np.random.uniform(0, 10)
    y[0]=np.random.uniform(0, 10)
    z[0]=np.random.uniform(0, 10)

    mu_z_ = lambda x_1, x_2: mu_z + np.dot(cov_12, np.linalg.inv(cov_22)).dot([x_1-mu_x, x_2-mu_y])
    mu_xy = lambda z: np.array([mu_x, mu_y]) + np.array(cov_12)/4.0*(z-mu_z)

    for i in xrange(N+burnin):
        muxy = mu_xy(z[i])
        x[i+1], y[i+1] = np.random.multivariate_normal(muxy, cov_xy)

        muz = mu_z_(x[i+1], y[i+1])
        z[i+1] = np.random.normal(muz, sig_z)

    return x[1+burnin:], y[1+burnin:], z[1+burnin:]
```

```python
x,y,z=Gibbs(10000)
```

### Question 2: Motif finding for extraterrestrial biology

You are a biologist who is analyzing some genetic material that astronauts collected in a nearby asteroid. The DNA of this extraterrestrial form of life is formed as a sequence of $n_B = 5$ nucleotide bases (unlike your DNA, that is made by combining only 4 nucleotides). You have named these extraterrestrial nucleotides 'F', 'R', 'L', 'A', and 'E'. NASA has given you a DNA sample consisting of $p=20$ sequences of DNA, each of them with a lenght $l=100$ nucleotide bases that you can find in <a href="Sequences.dat">this file</a>. Each row of the table is a DNA string. You are asked to find a particular substring (we call it a motif) of $k = 6$ consecutive nucleotides that appears to repeat itself very often in the DNA you were given, but that appears only once in each of the sequences. You quickly realize that the problem is equivalent to maximizing the joint probability distribution $p(x_1,x_2,...,x_p)$ in a p-dimensional space that the most "mutually similar" motif (in this case an identical sequence of $k$ elements) starts at positions $\left\{x_1,x_2,...,x_p\right\}$ of your set of $p$ sequences. 

Think about this in terms of Gibbs sampling. Maybe you can find the conditional probability of the hidden motif starting at position $j$ of your *i-th* string, given the current starting positions of the $k$-subsequence in all other strings. Let us help you out:

(a) Start by assigning random starting positions for the $k$-long motif in all sequences. Now exclude the first string and construct a matrix with dimension $(k+1)\times n_B$ where, for $j$ in $[1,k+1]$, element in column $j$ and row $i$ is the number of times that the *i-th* base is in position $j$, given the current starting points of the $p-1$ remaining sequences. Add an extra column (column 0) to account for the number of times that base the *i-th* appears in the background (not inside a motif). Now calculate a probability matrix:

$$
    P[i,j]= 
\begin{cases}
    \frac{C[i,j]+0.5}{p-1+0.5*n_B},& \text{           for } j \text{ in } [1,k+1]\\
    \frac{C[i,j]+0.5}{(l-k)*(p-1)+0.5*n_B},& \text{           for } j = 0
\end{cases}
$$

where the $C[i,j]$ are the number of counts in each case. The additional term of $0.5$ is a pseudocunt added to avoid infinities. 

(b) Think about how you can use this probability matrix to calculate the conditional probability $p(x_j|x_1,x_2,...,x_{j-1},x_{j+1},...,x_p)$) that the hidden motif starts in position $x_j$ of the first sequence. HINT: Given the probability of bases in different positions for all other sequences, what is the probability of having the observed nucleotide bases in the current sequence and current position?
 
(c) Construct a Gibbs sampling algorithm that, starting at random positions $\left\{x_1^{j},x_2^{j},...,x_p^{j}\right\}$ for all sequences, iterates this process over the $p=20$ different DNA sequences and generates an updated set of starting positions $\left\{x_1^{j+1},x_2^{j+1},...,x_p^{j+1}\right\}$ for the sequences, sampling from the conditional probabilities, until the updated positions converge. Provide the hidden motif and the starting positions of it in each of the sequences.

**(a)**

```python
with open('Sequences.dat','r') as f:
    DNA=[L.split('\n')[0] for L in f]
```

```python
k = 6
l = 100
p = 20
n_b = 5
match = {'A':0, 'E':1, 'F':2, 'L':3, 'R':4}

def ProbabilityMatrix(start_points, DNA, cur):
    P = np.zeros((n_b,k+1))
    C = np.zeros((n_b,k+1))
    for row in range(0,cur)+range(cur+1,p):
        tmp = np.zeros((n_b, k+1))
        start = start_points[row]
        
        for idx, nucleotide in enumerate(DNA[row][0:start]):
            tmp[match[nucleotide],0]+=1

        for idx, nucleotide in enumerate(DNA[row][start:start+k]):
            tmp[match[nucleotide],idx+1]+=1

        for idx, nucleotide in enumerate(DNA[row][start+k:l]):
            tmp[match[nucleotide],0]+=1

        # update value of C
        C += tmp

    # Calculate the probability matrix
    for i in xrange(n_b):
        for j in xrange(k+1):
            if j==0:
                P[i,j] = (C[i,j]+0.5)/((l-k)*(p-1)+0.5*n_b)
            else:
                P[i,j] = (C[i,j]+0.5)/((p-1)+0.5*n_b)

    return P
```

```python
start_points = np.random.randint(0,l-1-k,p)
P = ProbabilityMatrix(start_points, DNA, 0)
```

**(b)**

Suppose the motif starts in position $x_j$ of the first sequence is $\{n_1, n_2, n_3, n_4, n_5, n_6\}$. The set of other background nucleotides in this sequence is $n_\text{back}$ and the set of all nucleotides in this sequence is $n_\text{all}$. From problem (a), we can use the given positions for all other sequences to calculate prabability matrix $P$. The likelihood of this motif in position $x_j$ is 

$$\begin{align}
    p(x_j|x_1,x_2,...,x_{j-1},x_{j+1},...,x_p) 
    &= {P[n_1,1]P[n_2,2]P[n_3,3]P[n_4,4]P[n_5,5]P[n_6,6] \cdot \text{Product of Probability of other background nucleotides} }\\
    &= {P[n_1,1]P[n_2,2]P[n_3,3]P[n_4,4]P[n_5,5]P[n_6,6]\prod\limits_{n\in n_\text{back}} P[n,0]}\\
    &=\frac{P[n_1,1]P[n_2,2]P[n_3,3]P[n_4,4]P[n_5,5]P[n_6,6]}
    {P[n_1,0]P[n_2,0]P[n_3,0]P[n_4,0]P[n_5,0]P[n_6,0]}\prod\limits_{n\in n_\text{all}} P[n,0]
  \end{align}
$$

**(c)**

At the beginning, we randomly choose start position of every sequence as $\left\{x_1^{0},x_2^{0},...,x_p^{0}\right\}$

Assume we have start positions $\left\{x_1^{i+1},...x_{j-1}^{i+1}, x_j^{i},...,x_p^{i}\right\}$ now and we need to update $x_j$.

+ Step 1:

    Compute Probability Matrix $P$ based on sequences exclude $j^{th}$ sequence and $\left\{x_1^{i+1},...x_{j-1}^{i+1}, x_{j+1}^{i},...,x_p^{i}\right\}$


+ Step 2:

    From Problem (b), given start positions in other sequence, we can calculate the likelihood of each position in this sequence. To simplify, we just need to calculate  ratio of likelihood of motif vs background $\frac{P[n_1,1]P[n_2,2]P[n_3,3]P[n_4,4]P[n_5,5]P[n_6,6]}{P[n_1,0]P[n_2,0]P[n_3,0]P[n_4,0]P[n_5,0]P[n_6,0]}$ for each position because the other part is the same. And then I normalize this likelihood function over different positions and randomly choose one position $x_j^{i+1}$ according to the corresponding probability we computed. 


+ Step 3:

    And then we can update our start position as. $\left\{x_1^{i+1},...x_{j}^{i+1}, x_{j+1}^{i},...,x_p^{i}\right\}$



We can sequentially update start position for each sequence by repeating Step 2 and 3. Then we can have $\left\{x_1^{i+1},x_2^{i+1},...,x_p^{i+1}\right\}$

Repeat steps above and update start position sequentially until $\left\{x_1^{i},...,x_p^{i}\right\}==\left\{x_1^{i+1},...,x_p^{i+1}\right\}$

```python
def Gibbs_DNA(start_points, DNA):
    k = 6
    l = 100
    p = 20
    n_b = 5
    
    for cur in xrange(p):
        
        P = ProbabilityMatrix(start_points, DNA, cur)
        seq = DNA[cur]
        likelihood = np.full(l-k,1)
        for start in xrange(l-k):
            string = seq[start:start+k]
            for idx, letter in enumerate(string):
                likelihood[start] *= P[match[letter],idx+1]/P[match[letter],0]
        
        likelihood = likelihood/sum(likelihood)
        select = np.random.multinomial(1,likelihood)
        
        start_points[cur] = np.argmax(select)
    
    
    return start_points
```
