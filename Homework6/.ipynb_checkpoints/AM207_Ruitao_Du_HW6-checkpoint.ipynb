```python
def good_slice_sampler(pdf, x0, width, sample_size=5000, burnin=0, thin=0):
    
        # initialize the sampler
    y0 = np.random.uniform(low=0,high=pdf(x0))
    samples = []

    # randomly put the window
    u = np.random.uniform(0,1)
    # create an interval around the original point
    left = x0 - u*width
    right = left + width

    # slice sampling magic
    for i in range(sample_size + burnin):

        # select our interval
        while(y0 < pdf(left)): # left endpoint
            left = left - width

        while (y0 < pdf(right)): # right endpoint
            right = right + width

        while (True):
            # new proposal
            xstar  = np.random.uniform(low=left, high=right)

            # loop until we find the point inside the region
            if y0 < pdf(xstar):
                break
            else:
                # shrink interval
                if xstar < x0:
                    left = xstar
                if xstar > x0:
                    right = xstar

        # save the current sample
        samples.append(xstar)

        # the last sample is now the proposal
        x0 = xstar

        # get the new y0 for next step
        y0 = np.random.uniform(low=0, high=pdf(x0))
        
        # randomly put the window
        u = np.random.uniform(0,1)

        # reset the interval around x0
        left = x0 - u*width
        right = left + width


    return np.array(samples[burnin:])
```

In bad sampler, division is fixed. If we use a fixed division, our $x$ is always in the middle of the window so the window is less flexible. Sometimes we can go from $x$ to $x^*$, but we cannot go back to $x$.

I correct the bad sampler by changing fixed division to a uniformly random number from 0 to 1. By changing division, we are able to make sure that we have same probability from $x^*$ to $x$ and from $x$ to $x^*$. 

From histograms, we cannot tell the difference between slice samples and true samples. From qqplot, it is obvious that good slice sampler falls on the line. However, in qqplot of bad sampler, some samples are away from the line. Therefore, random division is important in slice sampling when we want to sample multimodal distribution.

$$ y_j \sim Poisson(n_j\theta_j) \\
 \theta_j \sim Gamma(\alpha, \beta)$$.

where $n_j$ is the total population of county j, $y_j$ is the number of kidney cancer deaths in county j and $\theta_j$ is the underlying death rate for county j.  A Poisson model makes sense because kidney cancer incidence should be a rare event (we hope!) and the Poisson is a good distribution for modeling rare events.  The gamma is a distribution with the right support that is conveniently the conjugate distribution for the Poisson. Before we were exposed to hierarchical modeling, we'd be inclined to use "full pooling" and choose one prior distribution for $\theta$ for all the counties.  Now that we have access to hierarchical modeling let's be a bit more sophisticated.  Let's define a hierarchical model where the $\alpha$ and $\beta$ are also drawn from gamma distributions as well.  Our full hierarchical model is defined below:
$$ y_j \sim Poisson(n_j\theta_j) \\
 \theta_j \sim Gamma(\alpha, \beta)\\
 \alpha \sim Gamma(a_{\alpha},b_{\alpha})\\
 \beta \sim Gamma(a_{\beta},b_{\beta})$$.

**(a)**

If prior is Gamma distribution
$$P(\lambda) \propto \frac{\beta^\alpha \lambda^{\alpha-1} e^{-\lambda\beta}}{\Gamma(\alpha)}$$

And likelihood is Poisson distribution
$$P(\{y_i\}_{i=1}^N|\lambda) \propto \prod\limits_{i=1}^N \frac{{\lambda}^{y_i}}{y_i!}e^{-\lambda}$$

The posterior would be

$$
\begin{align}
    P(\lambda|x) &\sim P(\lambda)P(\{y_i\}_{i=1}^N|\lambda)\\
    &\propto \frac{\beta^\alpha \lambda^{\alpha-1} e^{-\lambda\beta}}{\Gamma(\alpha)} \prod\limits_{i=1}^N \frac{{\lambda}^{y_i}}{y_i!}e^{-\lambda}\\
    &\propto \theta_j^{\alpha+\sum\limits_{n=1}^{N} y_i-1} e^{-\beta}\\
    &\propto Gamma(\alpha+\sum\limits_{n=1}^{N} y_i, \beta+N)
\end{align}
$$
So Gamma distribution is the conjugate prior of Poisson distribution. 

Our joint posterior is

And then we compute the joint posterior density.
$$
\begin{align}
    P(\lambda|x) &\sim P(\lambda)P(\{y_i\}_{i=1}^N|\lambda)\\
    &\propto Gamma(\alpha+\sum\limits_{n=1}^{N} y_i, \beta+N)
\end{align}
$$



**(b)**

The posterior distribution is
$$
\begin{align}
    p(\{\theta_i\}_{i=1}^N, \alpha, \beta|Y, n) 
    &\propto p(\alpha)p(\beta)\prod_{i=1}^N p(\theta_i|\alpha, \beta)\prod_{i=1}^N p(y_i|\theta, \alpha, \beta) \\
    &\propto p(\alpha)p(\beta)\prod_{i=1}^{N}\left\{\frac{\beta^\alpha}{\Gamma(\alpha)}\theta_i^{\alpha-1}e^{-\beta\theta_i}\frac{(n_i\theta_i)^{y_i}}{y_i!}e^{-n_i\theta_i}\right\}
\end{align}
$$

We can remove all terms that are not related to $\theta_i$ and calculate the conditional distribution for $\theta_i$. According to the conjugate priors

$$ p(\theta_i|\alpha, \beta, Y) \propto \text{Gamma}(\alpha+y_i, \beta+n_i) $$

We can remove all terms that are not related to $\beta$ and calculate the conditional distribution for $\beta$. According to the conjugate priors
$$ 
\begin{align}
    p(\beta|\{\theta_i\}_{i=1}^N, \alpha, Y)
    &\propto p(\beta) \beta^{N\alpha}\prod_{i=1}^{N}e^{-\beta\theta_i} \\
    &\propto \frac{{b_\beta}^{a_\beta}}{\Gamma(a_\beta)}\beta^{a_\beta-1}e^{-b_\beta\beta}\beta^{N\alpha}\prod_{i=1}^{N}e^{-\beta\theta_i} \\
    &\propto \text{Gamma}(a_\beta+N\alpha, b_\beta+\sum_{n=1}^N\theta_i)
\end{align}
$$

We can remove all terms that are not related to $\alpha$ and calculate the conditional distribution for $\alpha$. According to the conjugate priors

$$ p(\alpha|\{\theta_i\}_{i=1}^N, \beta, Y) \propto p(\alpha) \left[\frac{\beta^\alpha}{\Gamma(\alpha)}\right]^N\prod_{i=1}^{N}(\theta_i)^{\alpha-1}$$

```python
df_texas = df[df['state'] == "TEXAS"][['dc', 'pop', 'pct_mortality']]
y_i = df_texas['dc'].values
n_i = df_texas['pop'].values
theta_i = df_texas['pct_mortality'].values
```

```python
# set hyperparameters
a_alpha = 1.
b_alpha = 5.
a_beta = 15.
b_beta = 0.05
```

```python
numthetas = len(y_i)

# prior for the hyperparameters
log_prior_alpha = lambda a: np.log(scipy.stats.gamma.rvs(a, a_alpha, scale = 1./b_alpha))
log_prior_beta = lambda a: np.log(scipy.stats.gamma.rvs(a, a_beta, scale = 1./b_beta))

# sampling from conditional posterior for the theta
draw_thetas = lambda a,b: scipy.stats.gamma.rvs(a+y_i,scale = 1./(b+n_i))
   
# sampling beta from conditional posterior (pdf) for beta
def draw_beta(a,thetas):
    return scipy.stats.gamma.rvs(a_beta+numthetas*a, scale = (1./(b_beta+sum(theta_i))))

# conditional posterior (pdf) for alpha
def log_cond_alpha(a,b, thetas):
    return numthetas*(a*np.log(b) - scipy.special.gammaln(a)) + (a-1)*np.sum(np.log(thetas)) + log_prior_alpha(a)
  
    
# using the above pdfs we draw for alpha and beta in MH fashion
def draw_alpha(a,b,thetas,proposal_sd, accepts):
    astar = norm.rvs(a,proposal_sd)  # proposed a new alpha
    num = log_cond_alpha(astar,b, thetas) 
    den = log_cond_alpha(a,b, thetas)
    R=np.random.rand()
    #print num, den, R
    if (np.log(R) <= num - den) and (astar >0):#little built in rejection
        accepted=1
    else:
        accepted=0
    accepts['alpha'] = accepts['alpha'] + accepted
    if accepted:
        return astar
    else:
        return a
```

```python
burnin=20000
numtraces=200000
totaltraces=burnin+numtraces

alphas=np.empty(totaltraces)
betas=np.empty(totaltraces)
thetas=np.empty((totaltraces, numthetas))

#start positions
alphas[0]=1.
betas[0]=1.
thetas[0,:] = (y_i+0.5)/(n_i+0.5)

#draw_thetas(alphas[0], betas[0])
accepteds={'alpha':0}
```

```python
start = time.clock()

#Standard devation to draw alpha
prop_sd = .05 

for m in np.arange(1,totaltraces):
    alphas[m] = draw_alpha(alphas[m-1],betas[m-1],thetas[m-1,:], prop_sd, accepteds)
    betas[m] = draw_beta(alphas[m],thetas[m-1,:])
    thetas[m,:] = draw_thetas(alphas[m],betas[m])

print "Time Elapsed: ", time.clock()-start, " seconds."
```

Remove burnin samples and thinning.

```python
#remove burnin samples and thin
alphatrace=alphas[burnin+1:totaltraces:20]
betatrace=betas[burnin+1:totaltraces:20]
thetastrace=thetas[burnin+1:totaltraces:20,:]
print thetastrace.shape, alphatrace.shape
```
