## Question 1: Anxious Accountant

Assume that there are $N$ chain stores that are located at different places, labeled $1, 2, \ldots, N$; let $x_1,\ldots, x_N$ denote the locations. The total cost of transporting goods across all the stores every month is defined as
$$O = \sum_{i\neq j} f_{ij} d(x_i, x_j),$$
where $f_{ij}$ is the flow between store $i$ and store $j$. The store locations are designed such that the flow between store $i$ and store $j$ is inversely proportional to the distances between store $i$ and store $j$, i.e.
$$f_{ij} \propto 1/ d(x_i, x_j),$$
where $d(x_i, x_j)$ is the Euclidean distance between stores $i$ and $j$. Therefore the total cost of transporting goods is $O=\sum_{i\neq j} 1 = N(N-1)$, which results in a constant budget for transporting goods across stores every month. However, the careless accountant lost track of the labels of the stores, he is worried about being fired and thus decided not to tell the boss about this. His strategy is to randomly label the stores and use $f_{ij}$ as the flow between stores $i$ and $j$ in his labeling. Given a random permutation $\delta(\cdot)$ of the labels of the stores, the total cost becomes:
$$O = \sum_{i\neq j} f_{ij} d(x_{\delta(i)}, x_{\delta(j)}).$$
However, a simple calculation tells him that under many permutations, the total cost for transportation can be so large that a much higher budget for transportation is needed. How can the boss not discover such a sudden increase of the budget? Assume that the accountant comes to you and asks you to design a permutation such that the total cost is as small as possible, ideally to recover the true labels.
```python
# create a new permutation by swapping the number of n_swaps indexes randomly
def change_permute(permute, n_swaps):
    indices = range(permute.shape[0])
    permute_tmp = permute.copy()
    for i in range(n_swaps):
        swappedpermute = swapindex(permute_tmp)
        new_permute=swappedpermute.copy()

    return new_permute

def swapindex(permute):
    indices = range(permute.shape[0])
    # take two random indices to swap
    c1 = np.random.choice(indices)
    c2 = np.random.choice(indices)
    
    while c2 == c1:
        c2 = np.random.choice(indices) 
    
    # remember the indexes to swap
    tmp1 = permute[c1]
    tmp2 = permute[c2]
    
    # do the actual swapping
    changedpermute = permute.copy()
    changedpermute[c1] = tmp2
    changedpermute[c2] = tmp1
    
    return changedpermute
```

```python
def calculate_F(permute, locations):
    F = np.zeros((N,N))
    for i in xrange(len(F)-1):
        for j in range(i+1,len(F[0])):
            #print i,j, permute[i], permute[j]
            F[i,j] = 1./get_distance(locations[:,permute[i]], locations[:,permute[j]])
            F[j,i] = 1./get_distance(locations[:,permute[i]], locations[:,permute[j]])
    return F

def get_distance(loc1, loc2):
    return np.linalg.norm(loc1 - loc2)
```

```python
def calculate_cost(permute, locations, F):
    cost = 0
    for i in xrange(len(F)-1):
        for j in range(i+1,len(F[0])):
            cost += F[i,j]*get_distance(locations[:,permute[i]], locations[:,permute[j]])
    return cost
```

```python
def log_p(X, beta):
    res = 0.
    res += np.sum(X==Y1) + np.sum(X==Y2) + np.sum(X==Y3)
    res += beta*np.sum(X[1:,1:]==X[:-1,:-1])*2 
    res += beta*np.sum(X[1:,:]==X[:-1,:])*2 
    res += beta*np.sum(X[:,1:]==X[:,:-1])*2 
    res += beta*np.sum(X[1:,:-1]==X[:-1,:1])*2

    return res
```

```python
def proposal(x, num):
    I = np.random.randint(low=0, high=len(x), size=num)
    J = np.random.randint(low=0, high=len(x[0]), size=num)
    x_new = x.copy()
    for idx in xrange(num):
        x_new[I[idx], J[idx]] = 1-x_new[I[idx], J[idx]] 
        
    return x_new

# Metropolis Hasting
def Metropolis_hasting(theta0, logp, beta, proposal, var, num_samples, burnin=0, thinning=1, all_trace=True):
    
    N = num_samples    # number of samples to return 
    
    count = 0.    # number of accept samples
    
    theta = np.zeros((N*thinning+burnin+1,theta0.shape[0], theta0.shape[1]))
    theta[0] = theta0
    
    A = np.random.uniform(0.0, 1.0, N*thinning+burnin)
    
    prev_log_prob = logp(theta[0], beta)
    
    # begin iteration
    for i in xrange(N*thinning+burnin):
        theta_tmp = proposal(theta[i], var)
        log_prob = logp(theta_tmp, beta)
        #print log_prob - prev_log_prob, np.log(A[i])
        alpha = log_prob-prev_log_prob
        
        
        if(np.log(A[i])<alpha):   # Accept
            theta[i+1] = theta_tmp
            count+=1.
            prev_log_prob = log_prob

        else:            # Reject
            theta[i+1] = theta[i]
        
        #if not i%10000:
        #    print i
            
    print "The acceptance rate is " + str(float(count)/(N*thinning+burnin+1))+"."
    
    # if we just need to plot, I just return the mean of the samples
    if all_trace:    
        return theta[burnin+1::thinning]
    else:
        return np.mean(theta[burnin+1::thinning],axis=0)
```

```python
# log_p is the log target probability distribution, X0s is a list of initalizations for each chain
# Ts is a list of temperatures, n is the number of samples taken per sweep, N is the number of sweeps
def PT_simple(log_p, Ts, n, N, beta):
    
    num_chains = len(Ts)
    
    #create temperature-influenced distributions
    Ps = []
    for chain in xrange(num_chains):
        Ps.append(lambda x, beta_=beta, chain=chain: (log_p(x, beta_)/Ts[chain]))
    
    X0s = [np.random.randint(low=0, high=2, size=(len(Y1),len(Y1[0]))) for i in range(num_chains)]
    
    # store the last sample of each chain
    x=[ X0s[i] for i in range(num_chains)]
    
    # store the mean of the samples of each chain
    mean = [X0s[i] for i in range(num_chains)]
    
    for sweep in xrange(N):
        for chain in range(num_chains):
            #AUGMENTING MORE SAMPLES IN THE CHAIN
            sample = Metropolis_hasting(X0s[chain], Ps[chain], beta,proposal,3, n, burnin=1000, thinning=1)
            
            # I just use the last sample to compare with the other chain
            x[chain] = sample[-1]
            
            # add averge of samples to the mean
            if sweep==0:
                mean[chain] = np.mean(sample, axis=0)
            else:
                mean[chain] += np.mean(sample, axis=0)
 
        for chain in range(num_chains-1):
            x1 = x[chain]
            x2 = x[chain+1]
            
            # compare last sample of chain and chain+1 
            A = (Ps[chain](x2)+Ps[chain+1](x1)) - (Ps[chain](x1)+Ps[chain+1](x2)) # ACCEPTANCE PROBABILITY 
            if np.log(np.random.uniform()) < A:
                # accept 
                X0s[chain] = x2
                X0s[chain+1] = x1
            else:
                # reject 
                X0s[chain] = x1
                X0s[chain+1] = x2 

        if not sweep%10: print sweep
                
    # return the mean of each chain
    return mean
```
