### Question 1: Learn how to launch space rockets

In 1986, the space shuttle Challenger exploded during take off,
killing the seven astronauts aboard. It is believed that the
explosion was caused by the failure of an O-ring (a rubber ring that

In the file <a href="./files/chall.txt">chall.txt</a>, you will find temperature (in

stands for O-ring failure and 0 no failure.  We assume that the
observed temperatures are fixed and that, at temperature $t$, an
O-ring fails with probability $f(\theta_{1}+\theta_{2}t)$
conditionally on $\Theta = (\theta_1, \theta_2)$.

Here, $f(z) = 1/(1 + \exp(-z))$ is the logistic function.

Solution:

From each pair of data, we have temperatures $t$ and an indicator $k$ where $k=1$ means failure.

\begin{align}
P(\theta_1, \theta_2|D)&\propto P(D|\theta_1, \theta_2)P(\theta_1, \theta_2)\\
&\propto \prod\limits_{i=1}^{23} f(\theta_{1}+\theta_{2}t_i)^k (1-f(\theta_{1}+\theta_{2}t_i))^{k-1} \frac{1}{{2 \pi}} {\rm exp}\left[-\left(\frac{(\theta_1-26)^2}{2} + \frac{(\theta_2+0.4)^2}{2}\right) \right]
\end{align}

```python
# read in the data
chall = np.genfromtxt("files/chall.txt")
chall.shape
```

```python
#define logistic function
def f(theta_1, theta_2, t):
    return 1/(1+np.exp(-theta_1-theta_2*t))

# define prior
mu = [30, -0.3]
cov = [[30,0],[0,.3]]
rv = scipy.stats.multivariate_normal(mu, cov)
def prior(theta_1, theta_2):
    return rv.pdf([theta_1, theta_2])

# define posterior
def posterior(theta_1, theta_2):
    prob = f(theta_1, theta_2, chall[:,0])
    success = (1-chall[:,1])*(1-prob)
    failure = chall[:,1]*prob
    likelihood = np.prod(success+failure)
    return likelihood*prior(theta_1, theta_2)
```

```python
# true distribution
x_range, y_range = np.linspace(15, 40,100), np.linspace(-0.6, -0.2, 100)
k=0;j=0
pp=np.zeros((100,100))
for x1 in x_range:
    j=0
    for x2 in y_range:
        pp[k,j]=posterior(x1,x2)
        j +=1
    k +=1
plt.contourf(x_range, y_range, pp.T)
plt.colorbar()
plt.title('Posterior Contour Plot')
plt.xlabel('$\theta_1$')
plt.ylabel('$\theta_2$')
plt.show()

print "Maximum of posterior distribution is ",pp.max()
```

We want to increace the acceptance rate, so we narrow down the range of $\theta_1$ and $\theta_2$. Also, we change the maximum of p to $8\cdot 10^{-7}$ so as to increase the acceptance rate.

```python
N = 100000
theta_1_min, theta_1_max = 17  , 38
theta_2_min, theta_2_max = -.57, -.25
p_min, p_max =  0, 8e-7

begin = time.time()
samples = []

# sample at least N points
while(len(samples)<N):
    theta_1 = np.random.uniform(low=theta_1_min, high=theta_1_max) # sample theta_1
    theta_2 = np.random.uniform(low=theta_2_min, high=theta_2_max) # sample theta_2
    P = np.random.uniform(low=p_min, high=p_max) # sample p
    if P<posterior(theta_1, theta_2):  # accept
        if not len(samples)%10000: print len(samples), "samples....."  
        samples.append([theta_1, theta_2])

print "Time Elapsed:", time.time()-begin
```

### Question 2: Lensed galaxies

You are a renowned observational astronomer working on gravitatioanl lensing and you just got news about a source whose morphology appears distorted, most likely because there is a foreground source (an ensemble of mini black holes for which you know the mass and position) acting as a lense. Your gravitational lensing calculations indicate that the detected flux $F$ from the background source as a function of right ascencion ($x$) and declination ($y$) can be described by a modified Beale's function:

$F(x,y) = \exp\left[-\left(\frac{x^2}{2\sigma_x^2}+\frac{y^2}{2\sigma_y^2}\right)\right] \log (1.0+(1.5-x+xy)^2+(2.25-x+xy^2)^2+(2.625-x+xy^3)^2) $

You are interested in observing this source with the Hubble Space Telescope, and you want to simulate beforehand how photons will form the image on the Hubble detector. You realize that a good way to do this is by sampling F(x,y) with a Monte Carlo method.

(a) Using the following asymmetric function as a proposal distribution:

$q(x,y) = \frac{1}{{2 \pi \gamma_1\gamma_2}} {\rm exp}\left[-\left(\frac{(x-0.1)^2}{2 \gamma_1^2} + \frac{(y-0.1)^2}{2 \gamma_2^2}\right) \right] $ 

with $\gamma_1 = 1.0\beta$, $\gamma_2 = 1.5\beta$, and $\beta=1$ (x and y are the coordinates of the proposed step if we center the coordinate system in our current position.)

construct a Metropolis-Hastings algorithm to produce $N=100,000$ samples from $F(x,y)$, starting at $(x,y) = (5,-5)$ (normally you would start at a random position). Plot the results. As you are into it, think about the following: Could you make the algorithm simpler if your proposal function was symmetric? How? What is a good burn-in factor? (b) Experiment a little bit with $\beta$, by running your code with $\beta = 0.1,1.0,40.0$. Plot the accepted sample histories for each sample. What is the acceptance rate for each sample? What is the most efficient value for $\gamma$ to select the 100,000 samples?

(c) Now suppose that you are intereseted in the total flux arriving from your source, in the arbitrary units of the detector. Integrate the function $F(x,y)$ to find out. Run the integration 50 times. What is the variance of your result? Mention at least one method to reduce the variance of the integral that does not involve increasing the number of samples. Briefly explain how this method works.

```python
sigma_x, sigma_y = 10**.5, 10**.5

# define F(x, y) and q(x, y)
def F(x,y):
    return np.exp(-x**2/(2*sigma_x**2)-y**2/(2*sigma_y**2))*np.log(1.0+(1.5-x+x*y)**2+(2.25-x+x*y**2)**2+(2.625-x+x*y**3)**2)

# Metropolis Hasting
def metropolis(n_iterations, beta, burnin):

    gamma_1, gamma_2 = 1.0*beta, 1.5*beta
    mu = [0.1, 0.1]                       # mean of q
    cov = [[gamma_1**2, 0],[0, gamma_2**2]]     # covariance matrix of q
    
    # define q(x, y)
    q = lambda x,y: scipy.stats.multivariate_normal.pdf([x,y], mu, cov)
    
    N = n_iterations                  # number of iterations
    
    trace = np.zeros([2,N+1])
    trace[0][0], trace[1][0] = 5.0, -5.0 # initial position
    accept = [] # accepted pairs
    
    
    A = np.random.uniform(0.0, 1.0, N)
    
    # begin iteration
    for i in xrange(N):
        step1, step2 = scipy.stats.multivariate_normal.rvs(mu, cov).T
        q1 = q(step1, step2)
        q2 = q(-step1, -step2)
        tmp1, tmp2 = trace[0][i]+step1, trace[1][i]+step2
        alpha = min(1, (F(tmp1,tmp2)*q2)/(F(trace[0][i],trace[1][i])*q1))
        
        #print alpha, F(tmp1,tmp2), q2, F(trace[0][i],trace[1][i]), q2, step1, step2
        if(A[i]<alpha):   # Accept
            trace[0][i+1], trace[1][i+1] = tmp1, tmp2
            accept.append([tmp1, tmp2])
        else:            # Reject
            trace[0][i+1], trace[1][i+1] = trace[0][i], trace[1][i]
            
    return trace[:,burnin:], accept

# autocorr function
def autocorr(x):
   result = np.correlate(x, x, mode='full')
   result = result/result[result.size/2]
   return result[result.size/2:]
```

**(a)**

#####Could you make the algorithm simpler if your proposal function was symmetric?

Yes. In MH algorithm, Markov chain 
then moves towards $x^*$ with the acceptance probability 

$$
A(x^{(i)},x^*) = \min\left[1, \frac{p(x^*) \, q(x^{(i)}|x^*)}{p(x^{(i)}) q(x^*|x^{(i)}) } \right] 
$$

When proposal function was symmetric, which means $q(x^{(i)}|x^*)=q(x^*|x^{(i)})$, our acceptance probability will become

$$
A(x^{(i)},x^*) = \min\left[1, \frac{p(x^*)}{p(x^{(i)}) } \right] 
$$

### Question 3: Integrating nasty functions with MC

Suppose you want to evaluate the following integral which you have failed to solve analytically despite your weekend efforts:

$I=\int_{0}^{\infty} \frac{x^4\, \sin\left(\sqrt{\ln{(x+1)}}\right)e^{-x}}{2+(x-4)^2} \, dx $

As we have seen in class, an alternative for numerical integration of such a nasty equation is to use Monte Carlo methods. In particular, we can solve $\int_{0}^{\infty} f(x) \, dx$ if we can split the integrand as $f(x) = h(x)g(x)$, and then approximate the integral using:

$I = \int_{0}^{\infty} h(x)\,g(x) dx  \approx \frac{1}{N} \sum\limits_{i=1}^{N} h(x_i)$ where the values of $x$'s are now drawn from $g(x)$. 

Alternatively we can approximate the integral as 
$\hat{I} \approx \frac{1}{N} \sum\limits_{i=1}^{N} g(x_i) $
where the values of $x$'s are now drawn from $h(x)$.

(a) What should be the properties of $g(x)$ and $h(x)$ for our MC approximation to be as accurate as possible? Can our integrand be split as a product of such functions? What part would you use as $g(x)$ and why?

(b) Use $1/(2+(x-4)^2)$ as your $g(x)$ (remember that you need to normalize it). Implement a Metropolis-Hastings algorithm that has as a stationary distribution to sample from, $g(x)$.  Run the simulation 50 times for 150,000 points. Report the value of the integral $\hat{I}$ and the VAR(${\hat{I}}$).

(c) Find a better candidate for $g(x)$ and recalculate the integral. By how much did your variance improve?

Solution:

**(a)**

I think $g(x)$ should exist integral in the domain of the original integral, because we need to draw $x$ from $g(x)$. If the integral does not exist, we can not normalize it to a distribution. Actually, this is importance sampling, so the shape of ideal $g(x)$ is similar to $f(x)$.

In our integrand, we can seperate them is follow:
$$g(x)=x^4e^{-x}, h(x) = \frac{\sin\left(\sqrt{\ln{(x+1)}}\right)}{2+(x-4)^2} $$

**(b)**

we sample from $g(x) =1/(2+(x-4)^2)$ using the Metropolis algorithm with proposal 
distribution 

$$x^* \sim N(0, 1) $$

First, I normalize g(x).

\begin{align} \int_{0}^{\infty}g(x)dx &= \int_{0}^{\infty}\frac{1}{2+(x-4)^2}dx\\
&= {\frac{\sqrt{2}}{2}arctan(\frac{x}{\sqrt{2}}-2\sqrt{2})}|_0^{\infty}\\
&\approx 1.9811
\end{align}

$$\Rightarrow h(x) = \frac{1.9811 \cdot x^4\, \sin\left(\sqrt{\ln{(x+1)}}\right)}{e^x} $$

```python
g = lambda x: 1.0/(2.0+(x-4.0)**2.0)

# Metropolis Hasting
def metropolis(n_iterations, burnin):
    
    mu = 0     # mean of proposal distribution
    sd = 1     # standard deviation of proposal distribution
    
    # define q(x, y)
    q = lambda x: scipy.stats.normal.pdf(x, mu, sd**2)
    
    N = n_iterations                  # number of iterations
    
    trace = np.zeros(N+1)
    trace[0]= 1.0 # initial position
    accept = [] # accepted pairs
    
    
    A = np.random.uniform(0.0, 1.0, N)
    
    # begin iteration
    for i in xrange(N):
        x_star = np.random.normal(trace[i], sd)
        while  x_star <0 :  # MAKE SURE WE STAY WITHIN BOUNDS
            x_star = np.random.normal(trace[i], sd)
        alpha = min(1, g(x_star)/g(trace[i]))
        
        if(A[i]<alpha):   # Accept
            trace[i+1] = x_star
            accept.append(x_star)
        else:            # Reject
            trace[i+1] = trace[i]
            
    return trace[burnin:], accept
```

```python
h = lambda x: x**4.0*np.sin(np.log(x+1)**.5)*np.exp(-x)*1.9811

N_s = 50     # number of simulations
N_p = 150000 # number of points
Integral = np.zeros(N_s)
    
for i in xrange(N_s):
    trace, accept = metropolis(N_p, 1000)
    Integral[i] = np.mean(h(trace))
```

```python
plt.hist(Integral)
plt.title('Histogram of Simulated Integral')
plt.show()

print "The simulated value of integral is %f. The variance of integral is %f" % (np.mean(Integral),np.var(Integral))
```
