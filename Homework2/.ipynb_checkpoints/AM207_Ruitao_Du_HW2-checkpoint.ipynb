### Question 1:  My Lover's Got Humor (*Hozier*)

It is rumored that buried under the piled up snows brought to us by the Blizzardpocolypse of 2015 theres is a modest road known as the Boulevard of Broken Dreams (*Green Day*).  Along this road you'll find scattered the shattered hearts of humorists who are "the giggle at a funeral" (*Hozier*) and also of one Richard Sherman (*Seattle Seahawks*). Let's assume the broken hearts are distributed according to the function $\heartsuit(\theta) \sim \frac{ {\rm sin^{34}}\theta}{\theta^2}$ from $0< \theta< \infty$ and $\heartsuit(\theta )=0$ otherwise.  As a current humorist, it is your job to try to integrate $\heartsuit(\theta)$.

(a) Estimate your integral of $\heartsuit (\theta)$ using the standard Monte Carlo method with $N=100000$.  That is calculate the integral, 

$I=\int_{0}^{\infty} \frac{ {\rm sin^{34}}\theta d\theta}{\theta^2}$

Repeat the simulation 100 times and report the mean value of $I$ and its variance. (Hint: find a way to deal with that $\infty$ sign there)<br>

(b) Use importance sampling to evaluate the same integral.  Use the pdf of $\mathcal{N}(20,0.5)$ for your importance sampling distribution.  Why is this not a good choice? Try to choose a good importance sampling distribution and evaluate again.  Do you see a reduction in variance?  For both importance sampling distributions, run 100 simulations of 100,000 points. 

(c) Use a control variate to reduce the variance of the estimation of $I$ using $N=100000$ and 100 simulations as in part (a)

(d) Is it a good idea to use the antithesis method to reduce variance here?  Why or why not?  Could you use antithesis in combination with another method?  (We're only looking for your understanding of the methods in this problem.  No code required for this part).

(e) ** CHALLENGE/EXTRA CREDIT ** Use the best variance reduction method/combination of methods you can come up with.  Please restrict yourself to methods that we've covered in class and lab.  The top 10 solutions will get an additional bonus. 


Answer:

(a) We need to set an upper bound to simulate infinity. We know that when x goes to 1000, the integrand $\frac{sin^34(\theta)}{\theta^2}$ goes to smaller than $10^{-6}$, which is small. Therefore, I set the upper bound to 1000.

```python
x_min, x_max = 0, 1000
N_s = 100
N_mc = 100000
I = np.zeros(N_s)

for i in xrange(N_s):
    X = np.random.uniform(low=x_min, high=x_max, size=N_mc)
    Integrand = np.sin(X)**34/X**2
    I[i] = (x_max-x_min)*np.mean(Integrand)
```

```python
print "The mean value of I is %.8f and its variance is %.8f" % (np.mean(I), np.var(I))
```

(b)

```python
# range of integration
theta_max = np.inf
theta_min = 0

mu = 20
sigma = 0.5**.5
f = lambda x: np.sin(x)**34/x**2
p = lambda x: scipy.stats.norm.pdf(x, loc = mu, scale = sigma)
p_cdf = lambda x:  scipy.stats.norm.cdf(x, loc = mu, scale = sigma)
```

```python
N_s = 100
N_mc = 100000
I_normal = np.zeros(N_s)

for i in xrange(N_s):
    theta = np.random.normal(20, 0.5**.5, N_mc)
    theta = theta[ (theta<theta_max) & (theta>theta_min)]
    normal = p_cdf(theta_max)-p_cdf(theta_min)
    Integrand = f(theta)/p(theta)*normal
    I_normal[i] = np.mean(Integrand)
    
print "The mean value of I is %f and its variance is %f" % (np.mean(I_normal), np.var(I_normal))
```

```python
plt.figure(figsize=[10,6])

x = np.linspace(0, 40, 400)
plt.plot(x, scipy.stats.norm.pdf(x, loc=20, scale=0.5**.5), 'r-', lw=3, alpha=0.6, label='Normal pdf')
plt.plot(x, np.sin(x)**34/x**2, 'b-', lw=3, alpha=0.6, label='Integrand')
plt.xlabel('x')
plt.ylabel('probability')
plt.title(u'Normal vs ${sin^{34}(x)}/{x^2}$')
plt.legend()
plt.show()
```

$\mathcal{N}(20,0.5)$ for your importance sampling is not good because the shape of this distribution is not similar to the integrand. In this case, most of the samples are from the lower part of the integrand so we cannot get the correct answer. To have a better importance sampling, I have to choose a distribution with the shape of pdf of this distribution similar to the integrand. Therefore, I choose exponentia distribution $exp(0.4)$.

```python
# range of integration
theta_max = np.inf
theta_min = 0

lamb = 0.40
f = lambda x: np.sin(x)**34/x**2
p = lambda x: scipy.stats.expon.pdf(x, scale = 1/lamb)
p_cdf = lambda x:  scipy.stats.expon.cdf(x, scale = 1/lamb)
```

```python
N_s = 100
N_mc = 100000
I_exp = np.zeros(N_s)
I = np.zeros(N_s)

for i in xrange(N_s):
    X = np.random.uniform(low=x_min, high=x_max, size=N_mc)
    I[i] = (x_max-x_min)*np.mean(f(X))
    
    # Importance Sampling
    theta = scipy.stats.expon.rvs(scale = 1/lamb, size = N_mc)
    theta = theta[ (theta<theta_max) & (theta>theta_min)]
    normal = p_cdf(theta_max)-p_cdf(theta_min)
    Integrand = f(theta)/p(theta)*normal
    I_exp[i] = np.mean(Integrand)
    
    
plt.figure(figsize=[14,8])
plt.subplot(1,2,1)
x = np.linspace(0, 40, 400)
plt.plot(x, scipy.stats.expon.pdf(x, scale=1), 'r-', lw=3, alpha=0.6, label='Exponential pdf')
plt.plot(x, np.sin(x)**34/x**2, 'b-', lw=3, alpha=0.6, label='Integrand')
plt.xlabel('x')
plt.ylabel('probability')
plt.title('Distribution of $sin^34(x)/x^2$ and Exponential')
plt.legend()

plt.subplot(1,2,2)
plt.hist(I,30, histtype='step', color = 'b', lw =2, alpha = 0.6, label=u'Basic MC')
plt.hist(I_exp,30, histtype='step', color = 'r', lw = 2, alpha=0.6, label = u'Importance Sampling')
plt.xlabel('Estimated Integrand')
plt.ylabel('frequency')
plt.title('Histogram of $sin^34(x)/x^2$ and Exponential')
plt.legend()
plt.show()


print "The mean value of I is %.8f and its variance is %.8f" % (np.mean(I_exp), np.var(I_exp))
```

The variance of importance sampling with $exp(0.4)$ reduces variance to $0.00000973$. And from the histogram on the right, we can see that the variance reduces compared to the basic MC method.

**(e)**

First, we can deal with $\infty$ sign with changing range of integration. Let $\theta = \frac{t}{1-t}$, where $0<t<1$. We have
$$\begin{align} I&=\int_{0}^{\infty} \frac{ {\rm sin^{34}}\theta d\theta}{\theta^2}\\
&=\int_{0}^{1}\frac{ {\rm sin^{34}}\frac{t}{1-t} d\frac{t}{1-t}}{(\frac{t}{1-t})^2}\\
&=\int_{0}^{1}\frac{ {\rm sin^{34}}\frac{t}{1-t} dt}{t^2}
\end{align}$$
This change of range is equivalent to importance sampling.

```python
x_min, x_max = 0, 1
N_s = 100
N_mc = 100000
I_e = np.zeros(N_s)

for i in xrange(N_s):
    X = np.random.uniform(low=x_min, high=x_max, size=N_mc)
    Integrand = np.sin(X/(1-X))**34/X**2
    I_e[i] = (x_max-x_min)*np.mean(Integrand)
    

print "The mean value of I is %.8f and its variance is %.8f" % (np.mean(I_e), np.var(I_e))
```

### Question 2:  Bayesian Unbreak My Code (*Whitney Houston*)

A particular digital communications sytem is known to transmit messages using the **lover's alphabet** of  Love Me's (1s) and Love Me Nots (0s).  Super secret code breakers at the UK's GCHQ have discovered by reading all love missives sent by anyone in the world that the "letter" Love Me is sent with a certain probability p and that p is either $\frac{1}{3}$ or $\frac{2}{3}$.  A junior analyst at GCHQ (a Yank on loan from the NSA) is given $4$ observations as data -- { Love Me, Love Me, Love Me Not, Love Me}.  A Bayesian, he is charged with inferring p from his observations.  What will he determine? Did "she used to love me a lot?" (*Johnny Cash*) <br>

Please help our NSA guy by answering the following questions:
1. What is the likelihood here?
1. How do you define the prior?
1. Given the above, what is the posterior for the problem?  *Hint, we are looking for two numbers.*
1. Which p is more likely given this posterior?

Answer:

1.Likelihood:

From the data, we don't know whether the data is ordinal, so I assume it is not ordinal. Therefore, the likelihood is
$$P(D|p) = 4\cdot p^3(1-p)$$

2.Prior:

We just know there are two options of p so I set the prior as equal possibility of these two values.
$$P(p=\frac{1}{3}) = \frac{1}{2}, P(p=\frac{2}{3}) = \frac{1}{2}$$

3.Posterior:

$$\begin{align}
P(p=\frac{1}{3}|D) &\propto P(D|p=\frac{1}{3})*P(p=\frac{1}{3})\\
&\propto4\cdot \frac{1}{3^3}(1-\frac{1}{3}) \frac{1}{2}\\
&\propto\frac{4}{81}
\end{align}$$

$$\begin{align}
P(p=\frac{2}{3}|D) &\propto P(D|p=\frac{2}{3})*P(p=\frac{2}{3})\\
&\propto4\cdot \frac{2^3}{3^3}(1-\frac{2}{3}) \frac{1}{2}\\
&\propto\frac{16}{81}
\end{align}$$

Therefore, normalize the answer above we can have
$$P(p=\frac{1}{3}) = \frac{1}{5}, P(p=\frac{2}{3}) = \frac{4}{5}$$
