```python
# define logp(x)
def logp(x):
    return -0.5*np.log(2*np.pi)-x**2/2.0

def proposal(x):
    return np.random.normal(loc=x)


# Metropolis Hasting
def Metropolis(theta0, logp, proposal, n_iterations, burnin=0):
    
    N = n_iterations                  # number of iterations
    
    theta = np.zeros(N+1)
    accept = np.zeros(N+1)
    theta[0] = theta0
    
    A = np.random.uniform(0.0, 1.0, N)
    
    # begin iteration
    for i in xrange(N):
        theta_tmp = proposal(theta[i])
        alpha = min(0, logp(theta_tmp)-logp(theta[i]))
        
        
        if(np.log(A[i])<alpha):   # Accept
            theta[i+1] = theta_tmp
            accept[i+1] = 1

        else:            # Reject
            theta[i+1] = theta[i]
            
    return theta[burnin+1:], accept[burnin+1:]
```

```python
# Geweke tests
def Geweke(trace, intervals, length):
    # take two parts of the chain. 
    # subsample lenght 
    nsl=length
    z =np.empty(intervals)
    for k in np.arange(0, intervals):
        # beg of each sub samples
        bega=k*length
        begb = len(trace)/2 + k*length
        
        sub_trace_a = trace[bega:bega+nsl]
        sub_trace_b = trace[begb:begb+nsl]
        
        theta_a = np.mean(sub_trace_a)
        theta_b = np.mean(sub_trace_b)
        var_a  = np.var(sub_trace_a)
        var_b  = np.var(sub_trace_b)
        
        z[k] = (theta_a-theta_b)/np.sqrt(var_a + var_b)
    
    return z
```
