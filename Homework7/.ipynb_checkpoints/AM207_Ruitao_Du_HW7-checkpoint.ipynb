```python
# Geweke tests
def Geweke(trace, intervals, length):
    # take two parts of the chain. 
    # subsample lenght 
    nsl=length
    jump = int(0.9*len(trace)/(2*intervals))
    first = 0.1*len(trace)
    z =np.empty(intervals)
    for k in np.arange(0, intervals):
        # beg of each sub samples
        bega=first+k*jump
        begb = len(trace)/2 + k*jump
        
        sub_trace_a = trace[bega:bega+nsl]
        sub_trace_b = trace[begb:begb+nsl]
        
        theta_a = np.mean(sub_trace_a)
        theta_b = np.mean(sub_trace_b)
        rho_a, rho_b = 1.0, 1.0
        # only compute autocorrelation at lag 1-0.1*nsl.
        for i in xrange(int(0.1*nsl)):
            rho_a += 2*rhot(sub_trace_a, i+1)
            rho_b += 2*rhot(sub_trace_b, i+1)
            
        # estimate the variance roughly
        var_a  = np.var(sub_trace_a)*rho_a/length
        var_b  = np.var(sub_trace_b)*rho_b/length
        
        z[k] = (theta_a-theta_b)/np.sqrt( var_a + var_b)
    
    return z

# number of effective
def rhot(x, t):
    n = len(x)
    return np.corrcoef(x[0:(n-t)], x[t:n])[0,1]

def neff(x, T):
    n = len(x)
    rho = 1.0
    for i in xrange(T):
        rho += 2*rhot(x, i+1)
    return n/rho

# autocorrelation
def autocorr(x):
   result = np.correlate(x, x, mode='full')
   result = result/result[result.size/2]
   return result[result.size/2:]
```

```python
def proposal(x):
    return np.random.multivariate_normal(mean=x, cov=[[1,0],[0,1]])

# Metropolis Hasting
def Metropolis_hasting(theta0, logp, proposal, num_samples, burnin=0, thinning=10):
    
    N = num_samples    # number of samples to return 
    
    count = 0.    # number of accept samples
    
    theta = np.zeros((N*thinning+burnin+1,2))
    theta[0] = theta0
    
    A = np.random.uniform(0.0, 1.0, N*thinning+burnin)
    
    # begin iteration
    for i in xrange(N*thinning+burnin):
        theta_tmp = proposal(theta[i])
        alpha = min(0, logp(theta_tmp[0], theta_tmp[1])-logp(theta[i][0], theta[i][1]))
        
        
        if(np.log(A[i])<alpha):   # Accept
            theta[i+1] = theta_tmp
            count+=1.

        else:            # Reject
            theta[i+1] = theta[i]
            
    print "The acceptance rate is " + str(float(count)/(N*thinning+burnin+1))+"."
            
    return theta[burnin+1::thinning]
```

```python
# define function f, log_f and derivative of log_f
log_f = lambda x, y: -x**4-y**4+2*x*y-51*x**2-51*y**2+100*np.sqrt(x**2+y**2)

f = lambda x, y: np.exp(-x**4-y**4+2*x*y-51*x**2-51*y**2+100*np.sqrt(x**2+y**2))

dlog_f = lambda x: np.array([4*x[0]**3-2*x[1]+102*x[0]-100*x[0]/np.sqrt(x[0]**2+x[1]**2),
                             4*x[1]**3-2*x[0]+102*x[1]-100*x[1]/np.sqrt(x[0]**2+x[1]**2)])
```

```python
def HMC(q0, num_samples, burnin, dU, logf, M, epsilon, L, thinning):
    n = num_samples*thinning+burnin
    qold = q0
    Q = np.zeros((n, 2))
    r = np.random.random(n)
    count = 0
    
    for i in xrange(n):
        p = np.random.normal(size = 2, loc=0, scale=np.sqrt(M))
        pold = p
        q = qold    
        p -= epsilon*dU(q)/2
        
        for j in xrange(L-1):
            q += epsilon*p/M
            p -= epsilon*dU(q)
        
        q += epsilon*p/M
        p -= epsilon*dU(q)/2
        qnew = q
        pnew = -p
        p = logf(qnew[0], qnew[1]) - (pnew**2).sum()/(2.*M) - logf(qold[0], qold[1]) + (pold**2).sum()/(2.*M) 
        if np.log(r[i]) < p:
            qold = qnew
            count += 1
        Q[i,:] = qold
        
    print "The acceptance rate is " + str(float(count)/n)+"."
    return Q[burnin::thinning,:]
```

```python
num_samples = 10000
thinning=10
burnin = 5000
q0 = np.random.uniform(low=[-2,-2], high=[2,2])
epsilon = 0.05
L = 10
M = 0.5
Q = HMC(q0, num_samples, burnin, dlog_f, log_f, M, epsilon, L, thinning)
```

## Hint for likelihood functions
Define $\bar{X} = \frac{\sum_{i=1}^n x_i}{n}$, $\bar{Y} = \frac{\sum_{i=1}^n y_i}{n}$, $S_{X}^2 =\frac{\sum_{i=1}^n x_i^2}{n}$, $S_Y^2=\frac{\sum_{i=1}^n y_i^2}{n}$, $S_{XY} = \frac{\sum_{i=1}^n x_i y_i}{n}$. 

** normal noise**

If we assume $\epsilon_i$ are i.i.d. Gaussian instead of i.i.d. t, the likelihood function is given by

$$p(y_1,\ldots, y_n|\beta_0,\beta,\sigma^2)\propto \frac{1}{\sigma^n}\exp(-\frac{\sum_{i=1}^n (y_i-\beta_0-\beta x_i)^2}{2\sigma^2}),$$

which can be simplified as

$$p(y_1,\ldots, y_n|\beta_0,\beta,\sigma^2)\propto \frac{1}{\sigma^{n}}\exp(-\frac{n (\beta_0^2 + \beta^2 S_{X}^2 + 2\beta_0\beta \bar{X} - 2\beta_0 \bar{Y} - 2\beta S_{XY}+S_Y^2)}{2\sigma^2}).$$

** t distribution noise**

If we assume $\epsilon_i$ are i.i.d. t distribution with degree of freedom $\nu$, the likelihood function is given by

$$p(y_1,\ldots, y_n|\beta_0,\beta,\sigma^2)\propto \frac{1}{\sigma^n}\prod_{i=1}^n (1+\frac{(y_i-\beta_0-x_i\beta_1)^2}{\sigma^2\nu})^{-\frac{\nu+1}{2}}.$$

```python
S_x2 = np.mean(x*x)
S_y2 = np.mean(y*y)
S_xy = np.mean(x*y)
mu_x = np.mean(x)
mu_y = np.mean(y)
N = len(x)
tau = 1.
v = 2.0

def log_likelihood(beta_0, beta, sigma2):
    return -N*np.log(sigma2)/2. - np.sum((y-beta_0-beta*x)**2)/(2*sigma2)

def log_flat_prior(beta_0, beta, sigma2):
    return -np.log(sigma2)

def log_conjugate_prior(beta_0, beta, sigma2):
    log_p_beta0 = np.log(sp.stats.norm.pdf(beta_0, loc=0, scale=0.5))
    log_p_beta  = np.log(sp.stats.norm.pdf(beta, loc=0, scale=0.5))
    log_p_sigma2 = (v/2.)*np.log(tau**2.*v/2.)+(-v*tau**2./2./sigma2)- sp.special.gammaln(v/2.) - (1+v/2.)*np.log(sigma2)
    return log_p_beta0+log_p_beta+log_p_sigma2

def log_flat_posterior(beta_0, beta, sigma2):
    return log_flat_prior(beta_0, beta, sigma2)+log_likelihood(beta_0, beta, sigma2)

def log_conj_posterior(beta_0, beta, sigma2):
    return log_conjugate_prior(beta_0, beta, sigma2)+log_likelihood(beta_0, beta, sigma2)
```

```python
def log_t_likelihood(beta_0, beta, sigma2):
    return -(v+1)/2.*np.sum(np.log(1+(y-beta_0-x*beta)**2/(sigma2*v)))-N*np.log(sigma2)/2.
    
def log_flat_posterior_t(beta_0, beta, sigma2):
    return log_flat_prior(beta_0, beta, sigma2)+log_t_likelihood(beta_0, beta, sigma2)

def log_conj_posterior_t(beta_0, beta, sigma2):
    return log_conjugate_prior(beta_0, beta, sigma2)+log_t_likelihood(beta_0, beta, sigma2)
```
